```python
import pandas as pd
```

# Cleaning coulmns (year)

```python
car = car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

# Cleaning coulmns (Price)

```python
car['Price'].unique()
```

```python
car=car[car['Price']!="Ask For Price"]
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

# Cleaning coulmns (kms_driven)

```python
car['kms_driven']
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car['kms_driven'].unique()
```

```python
car['kms_driven']=car['kms_driven'].str.isnumeric()
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car.info()
```

# Cleaning coulmns (name)

```python
car['name']
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car['name']
```

```python
car.info()
```

# Cleaning coulmns (fuel_type)

```python
car = car[~car['fuel_type'].isna()]
```

```python
car.info()
```

```python
car.describe()
```

```python
car = car.reset_index(drop=True)
```

```python
car
```

```python
car = car[car['Price']<5e6]
```

```python
car = car.reset_index(drop = True)
```

```python
car
```

```python
car.to_csv('cleaned_car.csv')
```

# MODEL

```python
X = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['name','company','fuel_type','year']])
```

```python
 column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type','year']),remainder='passthrough')
```

```python
ohe.categories_
```

```python
lr = LinearRegression()
```

```python
ppl=make_pipeline(column_trans,lr)
```

```python
from sklearn import set_config
```

```python
set_config(display='diagram')
```

```python
ppl.fit(X_train,y_train)
```

```python
y_pred = ppl.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```
